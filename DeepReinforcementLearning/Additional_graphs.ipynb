```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from itertools import cycle, islice
import os
import cv2
import matplotlib.image as img
```

A method to save figures

```python
# Where to save the figures
PROJECT_ROOT_DIR = "."
IMAGES_PATH = os.path.join(PROJECT_ROOT_DIR, "images")
os.makedirs(IMAGES_PATH, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=100):
    path = os.path.join(IMAGES_PATH, fig_id + "." + fig_extension)
    print("Saving figure", fig_id)
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
df = pd.DataFrame({"A2C":[12000], "DQN":[1351], "DDQN":[783], "DDDQN":[726]})
colors = ['b', 'r', 'g', 'y']

plt.figure(figsize=(8,8))
plt.bar(x=df.columns, height=df.iloc[0,:], color=colors)
plt.title("Episodes to finish", size=15)
plt.legend()
plt.xlabel("Algorithm", size=12)
plt.ylabel("No. of Episodes", size=12)
save_fig("Episodes_to_finish")
plt.show()
```

```python
fignames = ["DQN","DDQN","DDDQN","A2C"]
imdir = "C:\\Users\\Daniel Kuknyo\\Documents\\ELTE\\2\\Artificial Intelligence in Processes and Automation\\Project\\images\\Rewards_"
nrows = 2
ncols = 2
imid = 0
fig = plt.figure(figsize=(10,10))
fig.suptitle("Rewards across learning episodes for all agents",size=20)
for i in range(nrows+ncols):
    ax = fig.add_subplot(nrows,ncols,i+1)
    imname = fignames[imid]
    ax.set_title(imname, size=12)
    ax.axis('off')
    im = img.imread(imdir+fignames[i]+".png")
    ax = plt.imshow(im)
    imid+=1
plt.tight_layout()
save_fig("Episodic_rewards")
plt.show()
```
